```python
import pandas as pd

vg_new_df = pd.read_csv('vgchartz-2024.csv')
vg_old_df = pd.read_csv('vgsales.csv')
add_on_df = pd.read_csv('game_sales_data.csv', encoding='ISO-8859-1')
```

```python
vg_new_df = vg_new_df.drop(['img', 'last_update'], axis= 1)
display(vg_new_df)
```

```python
vg_new_df = vg_new_df.rename(columns={
    'title': 'Name',
    'console' : 'Platform',
    'genre' : 'Genre',
    'publisher' : 'Publisher',
    'total_sales': 'Global_Sales',
    'na_sales' : 'NA_Sales',
    'jp_sales' : 'JP_Sales',
    'pal_sales' : 'EU_Sales',
    'other_sales' : 'Other_Sales',
    'critic_score' : 'User_Score'
})

display(vg_new_df)
```

```python
vg_new_df['release_date'] = pd.to_datetime(vg_new_df['release_date'])

# Extract the year and create a new column 'Year'
vg_new_df['Date'] = vg_new_df['release_date'].dt.year

# Optionally, drop the original 'release_date' column if not needed
vg_new_df = vg_new_df.drop(columns=['release_date'])

display(vg_new_df)
```

```python
display(vg_new_df)
```

```python
add_on_df = add_on_df.rename(columns={"Game_Name": "Name", "User Score": "User_Score", "Year" : "Date"})
display(add_on_df)
```

```python
vg_old_df = vg_old_df.rename(columns={
    "Year": "Date"
})
display(vg_old_df)
```

```python
merged_df = pd.merge(vg_old_df, add_on_df[['Name', 'Platform', 'User_Score', 'Date', 'Total_Shipped']], 
                     on=['Name', 'Platform', 'Date'], how='left')

merged_df = merged_df.drop(['Rank'], axis=1)


display(merged_df)
```

```python
vg_new_df['game_id'] = range(1, len(vg_new_df) + 1)
merged_df['game_id'] = range(len(vg_new_df) + 1, len(vg_new_df) + len(merged_df) + 1)

merged_df['game_id'] = merged_df['game_id'].astype(int)
vg_new_df['game_id'] = vg_new_df['game_id'].astype(int)



vg_new_df
```

```python
merged_df
```

```python
final_merged_df = pd.merge(merged_df, vg_new_df, on=['Name'], how='inner')
final_merged_df = final_merged_df.drop_duplicates(subset='Name', keep='first', inplace=False)
final_merged_df = final_merged_df.drop(columns=[col for col in final_merged_df.columns if col.endswith('_y')])
final_merged_df = final_merged_df.rename(columns=lambda x: x[:-2] if x.endswith('_x') else x)
final_merged_df = final_merged_df.fillna(0)
display(final_merged_df)
```

```python
final_merged_df.to_csv('final_vg_sales.csv')


# Viewing the columnns on the cvs file
print("CVS Columns")
for col in final_merged_df.columns:
    print(col)

print("\n\n")

print("CVS Full")
display(final_merged_df)
```
